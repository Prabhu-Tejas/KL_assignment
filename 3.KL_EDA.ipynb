```python
# Define price ranges and corresponding categories
price_ranges = [(0, 500000), (500001, 1000000), (1000001, 1500000), (1500001, 2000000), (2000001, float('inf'))]
categories = ['Copper', 'Bronze', 'Silver', 'Gold', 'Premium']

# Function to map price to category
def map_price_to_category(price):
    for i, (lower, upper) in enumerate(price_ranges):
        if lower <= price <= upper:
            return categories[i]
    return categories[-1]  # Default to 'Premium' for prices higher than the defined ranges

# Apply the mapping function to create a new column with categories
data['Price Category'] = data['Price'].apply(map_price_to_category)

# Replace 'Property Type' based on 'Price Category'
property_type_mapping = {
    'Copper': 'Copper',
    'Bronze': 'Bronze',
    'Silver': 'Silver',
    'Gold': 'Gold',
    'Premium': 'Premium'
}
data['Property Type'] = data['Price Category'].map(property_type_mapping)
```

```python
# Define price and size ranges and corresponding categories
price_ranges = [(0, 500000), (500001, 1000000), (1000001, 1500000), (1500001, 2000000), (2000001, float('inf'))]
size_ranges = [(0, 1000), (1001, 2000), (2001, 3000), (3001, 4000), (4001, float('inf'))]
categories = ['Copper', 'Bronze', 'Silver', 'Gold', 'Premium']

# Function to map price and size to category
def map_price_size_to_category(price, size):
    for i, (price_lower, price_upper) in enumerate(price_ranges):
        for j, (size_lower, size_upper) in enumerate(size_ranges):
            if price_lower <= price <= price_upper and size_lower <= size <= size_upper:
                return categories[i]
    return categories[-1]  # Default to 'Premium' for prices or sizes higher than the defined ranges

# Apply the mapping function to create a new column with categories
data['Price Category'] = data.apply(lambda row: map_price_size_to_category(row['Price'], row['Size']), axis=1)

# Replace 'Property Type' based on 'Price Category'
property_type_mapping = {
    'Copper': 'Copper',
    'Bronze': 'Bronze',
    'Silver': 'Silver',
    'Gold': 'Gold',
    'Premium': 'Premium'
}
data['Property Type'] = data['Price Category'].map(property_type_mapping)
```
