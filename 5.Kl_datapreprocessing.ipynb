```python
#importing data management, data visualisation libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import regex as re
```

```python
data= pd.read_csv('KL_FE_data.csv')
data
```

```python
# Select columns of type 'object'
object_columns = data.select_dtypes(include=['object'])

# Calculate the number of unique values in each column
unique_values = object_columns.nunique()

# Print the number of unique values for each column
print(unique_values)
```

```python
y= data['Price']
y
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler

# Copy the original DataFrame
data_encoded = data.copy()

# Initialize LabelEncoder, OneHotEncoder, and StandardScaler
label_encoder = LabelEncoder()
onehot_encoder = OneHotEncoder()
scaler = StandardScaler()

# Initialize list to store column names for scaling
numeric_columns = []

# Iterate over columns
for column in data.columns:
    # Check if column is of object dtype
    if data[column].dtype == 'object':
        # Check number of unique values
        unique_count = data[column].nunique()
        if unique_count == 2:
            # Apply label encoding for binary categorical columns
            data_encoded[column] = label_encoder.fit_transform(data[column])
        elif unique_count > 2:
            # Apply one-hot encoding for categorical columns with more than 2 unique values
            encoded_values = onehot_encoder.fit_transform(data[[column]]).toarray()
            encoded_df = pd.DataFrame(encoded_values, columns=[f'{column}_{val}' for val in onehot_encoder.categories_[0]])
            data_encoded.drop(column, axis=1, inplace=True)
            data_encoded = pd.concat([data_encoded, encoded_df], axis=1)
    else:
        # Add numeric columns for scaling
        numeric_columns.append(column)

```

```python
data_encoded.drop(columns='Price', inplace = True)
```

```python
data_encoded['Price']=y
```
