```python
#importing data management, data visualisation libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv('KL_clean_data.csv')
data
```

```python
data['Property Type Supergroup'].value_counts()
```

```python
# Calculate percentiles
percentiles = data['Price'].quantile([0, 0.3, 0.6, 0.9, 1])

# Define bins and labels
bins = [percentiles.iloc[0], percentiles.iloc[1], percentiles.iloc[2], percentiles.iloc[3], percentiles.iloc[4]]
labels = ['Bronze', 'Silver', 'Gold', 'Deluxe']

# Create the new column 'type' based on price ranges
data['type'] = pd.cut(data['Price'], bins=bins, labels=labels, include_lowest=True)
data
```

```python
data['type'].nunique()
```

```python
data['type'].value_counts()
```

```python
import pandas as pd

# Calculate percentiles for the 'size' column
percentiles = data['Size'].quantile([0, 0.3, 0.6, 0.9, 1])

# Define bins and labels for the 'size_type' column
bins = [percentiles.iloc[0], percentiles.iloc[1], percentiles.iloc[2], percentiles.iloc[3], percentiles.iloc[4]]
labels = ['Small', 'Medium', 'Large', 'Extra Large']

# Create the new column 'size_type' based on size ranges
data['size_type'] = pd.cut(data['Size'], bins=bins, labels=labels, include_lowest=True)
```

```python
data['size_type'].nunique()
```

```python
data['size_type'].value_counts()
```
